## 28. maketrans()

```python
# str.maketrans(x, y, z) 
# x → kaunse characters replace karne hain
# y → kis se replace karna hai
# z (optional) → kaunse characters delete karne hain
```

```python
table = str.maketrans("abc", "123")  #maketrans() ka use character mapping (translation table) banane ke liye hota hai, 

text = "apple banana cat"
print(text.translate(table))
```

```python
table = str.maketrans("", "", "aeiou")

text = "hello world"
print(text.translate(table))
```

```python
table = str.maketrans({
    "a": "@",
    "e": "3",
    "i": "1"
})

text = "apple"
print(text.translate(table))  #DICTIONARY FORM
```
